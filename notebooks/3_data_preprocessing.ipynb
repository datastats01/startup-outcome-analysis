## Data Preprocessing

```python
import pandas as pd

df = pd.read_csv("../data/startup_success_dataset.csv")
```

```python
# Data Cleaning
# Check missing values
df.isnull().sum()

# Check duplicates
df.duplicated().sum()

# Remove duplicates (if any)
df = df.drop_duplicates()
```

```python
# Feature Selection
# Target variable
y = df['outcome']

# Features
X = df.drop('outcome', axis=1)
```

```python
# Rescaling
from sklearn.preprocessing import StandardScaler

num_cols = [
    'team_size',
    'product_traction_users',
    'burn_rate_million',
    'revenue_million',
    'market_size_billion'
]

scaler = StandardScaler()

X[num_cols] = scaler.fit_transform(X[num_cols])
```

```python
# One-hot encoding
from sklearn.preprocessing import LabelEncoder

le_sector = LabelEncoder()
le_founder = LabelEncoder()
le_investor = LabelEncoder()

X['sector'] = le_sector.fit_transform(X['sector'])
X['founder_background'] = le_founder.fit_transform(X['founder_background'])
X['investor_type'] = le_investor.fit_transform(X['investor_type'])
```
